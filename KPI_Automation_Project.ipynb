```python
import pandas as pd

file_path = '/content/software_subscription_kpi_dataset.csv'
df_realistic = pd.read_csv(file_path)

total_revenue = df_realistic['SalesAmount'].sum()

average_marketing_spend = df_realistic['MarketingSpend'].mean()

total_leads = len(df_realistic)
converted_leads = df_realistic['LeadToCustomerConversion'].sum()
conversion_rate = (converted_leads / total_leads) * 100

total_marketing_spend = df_realistic['MarketingSpend'].sum()
marketing_roi = ((total_revenue - total_marketing_spend) / total_marketing_spend) * 100

print(f"Total Revenue: ${total_revenue:,.2f}")
print(f"Average Marketing Spend Per Transaction: ${average_marketing_spend:,.2f}")
print(f"Lead-to-Customer Conversion Rate: {conversion_rate:.2f}%")
print(f"Marketing ROI: {marketing_roi:.2f}%")
```

```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders

def send_email():
    sender_email = "muhammad.ahsan1325@gmail.com"
    receiver_email = "muhammad.ahsan1325@gmail.com"
    password = "password"


    msg = MIMEMultipart()
    msg['From'] = sender_email
    msg['To'] = receiver_email
    msg['Subject'] = "Weekly KPI Report"


    with open('kpi_report.txt', 'r') as f:
        body = f.read()
    msg.attach(MIMEText(body, 'plain'))


    for filename in ['total_revenue_by_region.png', 'total_revenue_by_product.png',
                     'conversion_rate_by_product.png', 'marketing_roi_by_product.png']:
        with open(filename, 'rb') as attachment:
            part = MIMEBase('application', 'octet-stream')
            part.set_payload(attachment.read())
            encoders.encode_base64(part)
            part.add_header(
                'Content-Disposition',
                f'attachment; filename= {filename}',
            )
            msg.attach(part)


    with smtplib.SMTP('smtp.gmail.com', 587) as server:
        server.starttls()
        server.login(sender_email, password)
        server.send_message(msg)


send_email()
```

```python
import schedule
import time

def generate_and_send_report():
    print("Task is running...")
    try:
        send_email()
        print("Email sent successfully!")
    except Exception as e:
        print(f"Error during email sending: {e}")


schedule.every(10).seconds.do(generate_and_send_report)


print(schedule.jobs)

while True:
    print("Checking for tasks...")
    schedule.run_pending()
    time.sleep(1)

```
